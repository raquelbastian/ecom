# EDA: Univariate, Bivariate, and Correlation Analysis

This notebook demonstrates how to perform:
- Univariate analysis (distribution plots)
- Bivariate analysis (scatter and box plots)
- Correlation analysis (heatmap)

```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset (update the path as needed)
df = pd.read_csv('amazon.csv')

# Display the first few rows
df.head()
```

## Univariate Analysis: Distribution Plots

Visualize the distribution of each numerical feature in the dataset.

```python
# List of numerical columns to analyze
num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

for col in num_cols:
    plt.figure(figsize=(6, 3))
    sns.histplot(df[col].dropna(), kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

## Bivariate Analysis: Relationships

Explore relationships between pairs of variables using scatter and box plots.

```python
# Scatter plot for two numerical variables (if at least two exist)
if len(num_cols) >= 2:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=df[num_cols[0]], y=df[num_cols[1]])
    plt.title(f'{num_cols[0]} vs {num_cols[1]}')
    plt.xlabel(num_cols[0])
    plt.ylabel(num_cols[1])
    plt.show()

# Boxplot for a numerical variable by a categorical variable
cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
if cat_cols and num_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[cat_cols[0]], y=df[num_cols[0]])
    plt.title(f'{num_cols[0]} by {cat_cols[0]}')
    plt.xlabel(cat_cols[0])
    plt.ylabel(num_cols[0])
    plt.xticks(rotation=45)
    plt.show()
```

## Correlation Analysis: Heatmap

Visualize the correlation matrix for numerical features.

```python
# Compute correlation matrix for numerical features
if num_cols and not df[num_cols].isnull().all().all():
    corr = df[num_cols].corr()
    if not corr.empty:
        plt.figure(figsize=(10, 8))
        sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', square=True)
        plt.title('Correlation Matrix')
        plt.show()
    else:
        print("Correlation matrix is empty. No numerical columns with valid data.")
else:
    print("No numerical columns with valid data for correlation analysis.")
```
