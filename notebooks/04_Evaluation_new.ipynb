```python
import pandas as pd
import numpy as np

def calculate_catalog_coverage_fixed(df, rec_func, n_queries=100, N=5):
    """
    Measures catalog coverage by ensuring the DataFrame is passed correctly.
    """
    total_unique_items = df['product_id'].nunique()
    recommended_items = set()
    
    # 1. Ensure we only sample from products that actually exist
    test_queries = df['product_id'].dropna().unique()
    sample_queries = np.random.choice(test_queries, size=min(n_queries, len(test_queries)), replace=False)
    
    print(f"Calculating coverage for {len(sample_queries)} products...")
    
    for pid in sample_queries:
        try:
            # FIX: Explicitly passing 'df' as the first argument
            recs = rec_func(df, pid, N=N)
            if recs is not None and not recs.empty:
                recommended_items.update(recs['product_id'].astype(str).tolist())
        except Exception as e:
            # Uncomment below to debug specific failures
            # print(f"Error for {pid}: {e}")
            continue
            
    coverage_ratio = len(recommended_items) / total_unique_items
    return coverage_ratio * 100, len(recommended_items), total_unique_items

# --- Optimized Weights ---
final_weights = {
    'basic_cosine': 0.0542, 'pca_features': 0.0743, 'content_tfidf': 0.2000,
    'content_pca': 0.0578, 'review_text': 0.0003, 'sentiment': 0.1864,
    'topic_lda': 0.0563, 'reviewer_overlap': 0.2247, 'knn_numeric': 0.0839,
    'svd_collaborative': 0.0621
}

# 2. Run the test with the corrected lambda
coverage_pct, unique_count, total_count = calculate_catalog_coverage_fixed(
    df, 
    lambda d, pid, N: get_weighted_hybrid_recommendations(d, pid, N=N, weights=final_weights),
    n_queries=150
)

print(f"\n--- CORRECTED CATALOG COVERAGE ---")
print(f"Total Products: {total_count}")
print(f"Unique Items Recommended: {unique_count}")
print(f"Coverage: {coverage_pct:.2f}%")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def evaluate_diversity_and_novelty(df, rec_func, n_queries=50, N=5):
    diversity_scores = []
    recommended_pids = []

    # Get feature matrix for diversity (using your PCA or TF-IDF features)
    # Here we use the precomputed X_knn as a proxy for structural similarity
    _, X_features = prepare_features_for_knn(df)
    
    sample_queries = df['product_id'].sample(n_queries).tolist()

    for pid in sample_queries:
        recs = rec_func(df, pid, N=N)
        if recs is not None and not recs.empty:
            p_ids = recs['product_id'].tolist()
            recommended_pids.extend(p_ids)
            
            # Calculate Intra-List Similarity
            indices = df[df['product_id'].isin(p_ids)].index.tolist()
            if len(indices) > 1:
                sim_matrix = cosine_similarity(X_features[indices])
                # Diversity = 1 - average similarity (excluding self-similarity)
                avg_sim = (sim_matrix.sum() - len(indices)) / (len(indices) * (len(indices) - 1))
                diversity_scores.append(1 - avg_sim)

    # Calculate Novelty
    # Novelty(p) = -log2(Probability(p)) where Probability is based on rating_count
    pop_counts = pd.to_numeric(df['rating_count'], errors='coerce').fillna(0) + 1
    total_pop = pop_counts.sum()
    
    novelty_scores = []
    for rid in recommended_pids:
        item_pop = df[df['product_id'] == rid]['rating_count'].iloc[0]
        prob = (float(item_pop) + 1) / total_pop
        novelty_scores.append(-np.log2(prob))

    return np.mean(diversity_scores), np.mean(novelty_scores)

# Run the evaluation
avg_div, avg_nov = evaluate_diversity_and_novelty(
    df, 
    lambda d, pid, N: get_weighted_hybrid_recommendations(d, pid, N=N, weights=final_weights)
)

print(f"Intra-List Diversity: {avg_div:.4f}") # Higher is more diverse
print(f"Mean Novelty Score: {avg_nov:.4f}")    # Higher is more 'surprising'
```
